### conceptos de POO (programacion orientada a objetos) en python

## el prompt IA gemini
lenguaje Python, crear la clase "animal" con los atributos "nombre ", "raza" crear el constructor animal, y la función ladrar(), correr(), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo, proporciona tu los datos

```python
# Definición de la clase
class Animal:
    # El constructor: aquí es donde nacen las características del objeto
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función (método) para ladrar
    def ladrar(self):
        print(f"{self.nombre} dice: ¡Guau, guau! 💩")

    # Función (método) para correr
    def correr(self):
        print(f"{self.nombre} está corriendo velozmente. 👯‍♀️")
    def morder(self):
        print(f"{self.nombre} está mordiendo a la vecina 🧞‍♂️")

# --- Creación de los objetos (Instanciación) ---

# Objeto 1: Pitbull
pitbull = Animal("Rocky", "Pitbull")

# Objeto 2: Chihuahua
chihuahua = Animal("Chispa", "Chihuahua")
# creamos el objeto para perros callejeros
perrocallejero = Animal("solovino", "chafa")
# --- Uso de los objetos ---

print(f"Perro 1: {pitbull.nombre}, Raza: {pitbull.raza}")
pitbull.ladrar()
pitbull.correr()

print("-" * 20)

print(f"Perro 2: {chihuahua.nombre}, Raza: {chihuahua.raza}")
chihuahua.ladrar()
chihuahua.correr()

print("-" * 20)

print(f"Perro 3: {perrocallejero.nombre}, Raza: {perrocallejero.raza}")
perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.morder()
print("programa creado por Santiago pacheco NL = 43")
```
